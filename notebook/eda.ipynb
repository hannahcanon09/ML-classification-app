### **EDA**

```python
# Function to explore the dataset
def explore_data (df):
    num_rows=df.shape[0]
    num_features=df.shape[1]
    data_types=df.dtypes
    
    print(f'Number of rows: {num_rows}')
    print (f'Number of features: {num_features}')
    print(f'\nData Types:')
    return data_types

explore_data(df)
```

```python
# Function to perform EDA
def perform_eda(df):
    duplicates=df.duplicated().sum()
    missing_values=df.isna().sum()

    print (f'Number of duplicate rows: {duplicates}')
    print (f'\nNumber of missing values: {missing_values}')

perform_eda(df)
```

```python
# Summary statistics for each feature 
statistics=df.describe().T
print(f'Summary Satistics:')
statistics
```

```python
# Function to check all unique values and anomalies in the dataset
def check_column_values (df):
    print (f'Check all columns for anomalous values \n')

    # Loop  through each column and return all unique values for each column
    for column_name in df.columns:
        print (f'Unique values in: {column_name}')
        print (df[column_name].unique())
        print ('\n')

check_column_values(df)
```

### **Visualizations**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Format all figures in dark mode
plt.style.use('dark_background')

# Create correlation heatmap
def plot_correlation(df):
    fig, ax=plt.subplots(figsize=(15,5))

    # Define  the mask for the upper triangle (to avoid duplicate values) and annotate it with colorbar as heatmap
    mask=np.triu(np.ones_like(df.corr(), dtype=bool))

    # Plot the heatmap
    sns.heatmap(
        df.corr().round(2),
        ax=ax, 
        mask=mask, 
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    ax.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=10)
    plt.show()

plot_correlation(df)
```

```python
# Correlation heatmap as ranked
def ranked_correlation_heatmap(df, target_column):
    plt.figure(figsize=(4,8))
    heatmap=sns.heatmap(
        df.corr()[[target_column]].sort_values(by=target_column, ascending=False),
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    heatmap.set_title('Features Correlating with Satisfied')
    plt.show()

ranked_correlation_heatmap(df, target_column='satisfied')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create histogram and boxplot
def visualize_data(df, column_name):
    fig, axes=plt.subplots(
        figsize=(10,5),
        nrows=2,
        sharex=False
    )
    sns.set_style('dark')

    # Histogram plot
    sns.histplot(
        df[column_name],
        ax=axes[0],
        bins='auto',
        kde=True
    )

    axes[0].tick_params(axis='x', rotation=45)

    # Boxplot
    sns.boxplot(
        data=df,
        x=column_name,
        ax=axes[1]
    )

    plt.tight_layout()
    plt.show()

visualize_data(df, column_name='Flight Distance')
```
